# ICM Explorer

Interactive ICM equity and bubble factor analysis.

```python
import numpy as np
from ipywidgets import interactive, HBox, VBox, Output, FloatSlider
from IPython.display import display

from speculator_ev_engine.poker.icm import malmuth_harville_icm, chip_chop_icm, bubble_factor
from speculator_ev_engine.ui.core.charts import icm_equity_by_stack, bubble_factor_heatmap
```

```python
out = Output()

def update_icm(s1, s2, s3, blend):
    with out:
        out.clear_output(wait=True)
        stacks = np.array([s1, s2, s3])
        payouts = np.array([500.0, 300.0, 200.0])
        if blend >= 1.0:
            result = malmuth_harville_icm(stacks, payouts)
        else:
            result = chip_chop_icm(stacks, payouts, blend_weight=blend)
        fig = icm_equity_by_stack(result.equities, stacks, result.total_prize_pool)
        display(fig)
        plt.close(fig)

s1_s = FloatSlider(value=5000, min=1000, max=10000, step=100, description='Stack 1')
s2_s = FloatSlider(value=3000, min=1000, max=10000, step=100, description='Stack 2')
s3_s = FloatSlider(value=2000, min=1000, max=10000, step=100, description='Stack 3')
blend_s = FloatSlider(value=1.0, min=0.0, max=1.0, step=0.05, description='Blend')
w = interactive(update_icm, s1=s1_s, s2=s2_s, s3=s3_s, blend=blend_s)
display(VBox([HBox([s1_s, s2_s, s3_s, blend_s]), out]))
```
